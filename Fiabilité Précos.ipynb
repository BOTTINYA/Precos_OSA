```python
import pandas as pd
from time import time

import seaborn
import matplotlib.pyplot as plt

from google.cloud import bigquery
import google.datalab.bigquery as bq

import google.auth
```

Définition du client bigquery pour aller sourcer les données et requête

```python
credentials, project = google.auth.default(scopes=[
    'https://www.googleapis.com/auth/drive',
    'https://www.googleapis.com/auth/bigquery',
])

#client = bigquery.Client(project='project', credentials=credentials)
client = bigquery.Client(project='osa-2019', credentials=credentials)



sql = """
      SELECT DISTINCT * FROM `osa-2019.Precos_Enseignes.PrecosBrutesVSRetouchees`
      """ 
            

start_time = time()

df = client.query(sql).to_dataframe()         #Interrogation de BigQuery 

print('Querying and loading time = {:0.2f} s '.format(time() - start_time))
print('Request finished\n')
```
